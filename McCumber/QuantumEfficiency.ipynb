### Theoretical decay rates from simulations using chip parameters

```python
# Load theoretical decay rates vs n and plot
data = np.load('./t2_vs_n.npz')
n = data['n']
fp = data['spe']

# Plot
f, ax = plt.subplots()
ax.plot(n, fp, '.-')
ax.set_xlabel('n')
ax.set_ylabel('$\Gamma / \Gamma_0$')
```

```python
# REF: http://scipy-cookbook.readthedocs.io/items/FittingData.html
#
# Model for the observed decay rate in terms of non-radiative, radiative
# and average purcell factor (fp) over layer.
#
#  W_tot = W_nr + W_rad * fp(n)
#

# define our (line) fitting function
fitfunc = lambda p, x: p[0] + p[1] * x
errfunc = lambda p, x, y, err: (y - fitfunc(p, x)) / err

pinit = [40, 40]
out = optimize.leastsq(errfunc, pinit, args=(fp, ydata, yerr), full_output=1)

pfinal = out[0]
covar = out[1]

W_nr = pfinal[0]
W_rad = pfinal[1]

set_colour_map(3)
f, ax = plt.subplots()
ax.plot(n, fitfunc(pfinal, fp), '.-', label='Theoretical (W_nr = {0:.2g}, W_rad = {1:.2g})'.format(W_nr, W_rad))
ax.errorbar(xdata, ydata, yerr=yerr, label='Experimental')
ax.set_xlabel('n')
ax.set_ylabel('PL decay rate (s$^{-1}$)')
plt.legend(loc='best')

# Calculate QE from fitted params
QE = lambda W_rad, W_nr: W_rad / (W_rad + W_nr)
print('Quantum efficiency is {:.2f} %'.format(100*QE(W_rad, W_nr)))
```

# Creatore Method

Solve:
$$
\begin{bmatrix}
    \Gamma^a \\
    \Gamma^b \\
\end{bmatrix}
=
\begin{bmatrix}
            1 & 1\\
    \beta^b_a & 1 \\
\end{bmatrix}
\begin{bmatrix}
    \Gamma^a_{rad} \\
    \Gamma^a_{nrad} \\
\end{bmatrix}
$$

where $\Gamma^{a,b}=1/\tau_{a,b}$ is the measured SE rate.

```python
beta = 1.076342  # Theoretical SE ratio (Gamma_rad: b/a=water/air)
matrix = np.array([[1, 1], [beta, 1]])
matrix
```

```python
# LHS array of measured lifetimes (inverse decay rates)
tau_a = 12.25e-3  # Air
tau_b = 12.15e-3  # Water

lhs = np.array([[1/tau_a], [1/tau_b]])
lhs
```

```python
# Invert matrix
det  = (1 - beta)
inv = 1 / det * np.array([[1, -1], [-beta, 1]])
```

```python
# Solve radiative and non-radiative decay rates
rhs = inv @ lhs
rad = rhs[0][0]
nrad = rhs[1][0]

print('Radiative decay rate    : {:.3e}'.format(rad))
print('Non-radiative decay rate: {:.3e}'.format(nrad))
```

Quantum Efficiency:

$$
q = \frac{\Gamma_{rad}}{\Gamma_{rad} + \Gamma_{nrad}} = \frac{\tau_{mes}}{\tau_{rad}}
$$

```python
q = rad / (rad + nrad)
print('Quantum Efficiency is {:.3f}'.format(q))
```
