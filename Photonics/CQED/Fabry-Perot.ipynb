```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Fabry-Perot Interferometer Functions
# Assuming no absorption or scattering loss within the cavity.

from scipy.constants import pi, c
from numpy import sin

def transmission(f, phase):
    return 1/(1+(4*f**2 / pi**2)*sin(phase/2)**2)
              
def phase_shift(n, L_cav, lam_vac):
    return 4*pi*n*L_cav/lam_vac

def finesse(r1, r2):
    """Finesse of a cavity with two mirrors.
    """
    num = pi*(r1*r2)**0.25
    den = 1 - (r1*r2)**0.5
    return num/den

def quality_factor(omega, delta_omega):
    return omega/delta_omega

def t_cav(n, L_cav, r):
    return (n*L_cav) / (c*(1-r))
```

```python
n = 1.6
L_cav = 1.5E-6
lam_vac = 1.54E-6
r1 = 0.99
r2 = 0.04

f = finesse(r1, r2)

phase = np.linspace(-2*pi-3, 2*pi+3, 1000)
t = transmission(f, phase)

plt.plot(phase, t)
plt.xticks([-2*pi, 0, 2*pi], [r'$2\pi (m-1)$', r'$2\pi m$', r'$2\pi (m+1)$'])
plt.ylabel('Cavity transmission')
plt.xlabel('Round-trip phase shift')
```

```python
lam_vac = np.linspace(1E-6, 2E-6, 1000)

phase = phase_shift(n, L_cav, lam_vac)
t = transmission(f, phase)

plt.plot(lam_vac*1E6, t)
plt.ylabel('Cavity transmission')
plt.xlabel('Vacuum Wavelength (um)')

print('Mean transmission {0:.2%} '.format(np.mean(t)))
```
