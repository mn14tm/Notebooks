```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 5.0)

from numpy import exp
```

# Evaluate CRLB(tau)

```python
tau = 10  # Luminescence lifetime
T = tau/0.5  # Detection channel width
M = np.linspace(1*tau/T,100*tau/T,num=5000)  # Detection channels
```

```python
def CRLB(tau,T,M):
    a = (T**2/tau**2)
    b = exp(T/tau)/(exp(T/tau)-1)**2
    c = (M**2*exp(M*T/tau)) / ((exp(M*T/tau)-1)**2)

    return 1/(a*(b-c))
```

```python
# log x and y axis
plt.subplot(111)
plt.loglog(M*T/tau, CRLB(tau,T,M),'-',label='T='++'tau')
plt.grid(True, which="both")
plt.xlabel('MT/tau')
plt.ylabel('EN*CRLB(tau)/tau**2')
plt.ylim(1,100)
plt.legend(loc='best')
```
