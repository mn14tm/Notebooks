```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Chapter 4 - NumPy Basics
## Example: Random Walks

```python
import random
position = 0
walk = [position]
steps = 1000
for i in range(steps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk)
```

```python
# Alternatively can be evaluated as an array expression
draws = np.random.randint(0,2, size=steps)  # Create array of 0s and 1s
steps = np.where(draws > 0, 1, -1)          # Convert 0s to -1
```

```python
walk = steps.cumsum()
```

```python
plt.plot(walk)
```

```python
# Can extract stats from this numpy array object
walk.min()
```

```python
walk.max()
```

```python
# Find step at which the walk trajectory FIRST reaches 10 steps away from start
(np.abs(walk) >= 10).argmax()  # Argmax finds first occurence of true in a boolean array
```

```python
# Simulating many random walks at once is easy! Pass random function a 2-tuple as follows:
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))  # 2D array of 0 or 1
```

```python
steps = np.where(draws > 0, 1, -1)
```

```python
walks = steps.cumsum(1) # Along axis 1 (columns)
```

```python
walks.max()
```

```python
walks.min()
```

```python
# Compute min crossing time to 30 or -30 (note not all 5000 walks will cross 30!)
hits30 = (np.abs(walks) >= 30).any(1)
```

```python
hits30
```

```python
hits30.sum()  # Number that hit 30 or -30
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(1)  # Use the hits30 to only select walks that do cross 30 or -30
```

```python
crossing_times.mean()
```

```python
# Can also experiemnt with other step distributions
steps = np.random.normal(loc=0, scale=0.25, size=(nwalks,nsteps))  # Normal dist.
```
