# Gasket flow cell design coefficient generator

Based upon icell design in http://etheses.whiterose.ac.uk/7765/ (pg 101-102)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
a = 6.5  # Half-length between inlet and outlet ports
b = 3.5  # Half-width of flow cell (max length perpendicular to flow direction)
```

```python
c1 = -2*b/(a**3)
c2 = 3*b/(a**2)
print('c1={0:.6} and c2={1:.6}'.format(c1, c2))
```

```python
c2
```

```python
x = np.linspace(0,a)
y = c1*x**3 + c2*x**2 + 0.5
plt.plot(x,y)
plt.ylim(0, max(y)+0.5)
```
