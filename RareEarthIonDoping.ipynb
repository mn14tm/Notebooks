```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

# Calculate average ion spacing

```python
nm3_per_cm3 = 1E21
rho = 1E19  # Ion doping concentration [ion/cm3]
v = nm3_per_cm3 / rho  # Volume per ion [nm3/ion]
r = pow(3/(4*np.pi) * v, 1/3)  # Average radius of ion assuming a sphererical shape

print('Doping concentration: {:g} ions/cm3'.format(rho))
print('Average separation between Er ions: {:.2f} nm'.format(2*r))
```

```python
nm3_per_cm3 = 1E21
rho = np.logspace(18, 23, num=1000, endpoint=False) # num=22-16, 
v = nm3_per_cm3 / rho  # Volume per ion [nm3/ion]
r = pow(3/(4*np.pi) * v, 1/3)  # Average radius of ion assuming a sphererical shape

fig,ax = plt.subplots()
ax.plot(rho, r)
ax.set_xscale('log')
ax.set_xlabel(r'$\rho$ [ion/cm$^3$]')
ax.set_ylabel(r'd$_{ion-ion}$ [nm]')
# fig.savefig('ion_spacing_high.png', dip=900)
```
